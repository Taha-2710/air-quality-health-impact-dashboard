```python
df['date']=pd.to_datetime(df['date'], format='%Y-%m-%d')
```

```python
def outliers(df,columns,factor=1.5):
    df1=df.copy()
    for x in columns:
        Q1=df[x].quantile(0.25)
        Q3=df[x].quantile(0.75)
        IQR=Q3-Q1
        Lower_bound=Q1-1.5*IQR
        Upper_bound=Q3+1.5*IQR
        df1=df1[(df1[x]>=Lower_bound)&(df1[x]<=Upper_bound)]
    return df1

df_clean=outliers(df,columns=[
    'aqi',
    'pm2_5',
    'pm10',
    'no2',
    'o3',
    'temperature',
    'humidity',
    'hospital_admissions',
    'hospital_capacity'
])
df_clean.describe()
```

```python
df_clean['Year']=df_clean['date'].dt.year
```

```python
df_clean=df_clean[df_clean['Year']<=2024]
```
